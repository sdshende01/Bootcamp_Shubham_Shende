```python

import pandas as pd
import numpy as np
import os

DATA_PATH = "starter_data.csv"  # adjust if running elsewhere
df = pd.read_csv(DATA_PATH, parse_dates=["Date"])
print("Shape:", df.shape)
df.head()
```

```python

# Basic info and missing counts
display(df.info())
print("\nMissing counts by column:")
print(df.isna().sum())

# Numeric summary
display(df.describe())

# Categorical overview
cat_cols = df.select_dtypes(include=["object", "category"]).columns.tolist()
print("\nCategorical columns:", cat_cols)
for c in cat_cols:
    print(f"\nValue counts for {c}:")
    print(df[c].value_counts(dropna=False))
```

```python

import matplotlib.pyplot as plt

# Histogram: Value
df['Value'].dropna().hist(bins=30)
plt.title("Distribution of Value")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()

# Histogram: Amount
df['Amount'].dropna().hist(bins=30)
plt.title("Distribution of Amount")
plt.xlabel("Amount")
plt.ylabel("Frequency")
plt.show()

# Boxplot: Amount by Category
# Using matplotlib directly for a by-category boxplot
cats = sorted(df['Category'].dropna().unique())
data_by_cat = [df.loc[df['Category']==cat, 'Amount'].dropna().values for cat in cats]
plt.boxplot(data_by_cat, labels=cats)
plt.title("Amount by Category")
plt.xlabel("Category")
plt.ylabel("Amount")
plt.show()
```

## 4. Bivariate Visuals

```python

# Scatter: Value vs Amount (drop missing)
tmp = df[['Value','Amount']].dropna()
plt.scatter(tmp['Value'], tmp['Amount'])
plt.title("Value vs Amount")
plt.xlabel("Value")
plt.ylabel("Amount")
plt.show()

# Line: Date vs Value (sorted by date)
df_sorted = df.sort_values('Date')
plt.plot(df_sorted['Date'], df_sorted['Value'])
plt.title("Value over Time")
plt.xlabel("Date")
plt.ylabel("Value")
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

```python

# Correlation for numeric columns
num_df = df.select_dtypes(include=['number'])
corr = num_df.corr(numeric_only=True)

import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()
plt.imshow(corr.values, interpolation='nearest')
plt.title("Correlation Heatmap")
plt.xticks(range(len(num_df.columns)), num_df.columns, rotation=45)
plt.yticks(range(len(num_df.columns)), num_df.columns)
plt.colorbar()
plt.tight_layout()
plt.show()
```


**Findings, Risks, and Assumptions**

- Skew: Amount appears right-skewed (long tail); consider log-transform for modeling.
- Outliers: Boxplot by Category suggests potential outliers in Amount.
- Seasonality/Trend:Value over time shows visible structure.
- Missingness: Some missing values in Value and Amount.

**Assumptions & Risks:**  
- Assuming missing at random (may be wrong).  
- Assuming categories are comparable groups; imbalance could distort group-level comparisons.
